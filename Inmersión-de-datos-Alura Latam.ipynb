```python
import pandas as pd
inmuebles = pd.read_csv("/content/drive/MyDrive/Copia de inmuebles_bogota.csv")
inmuebles.head()
```

```python
columnas = {"Baños" : "Banos","Área" : "Area"}
inmuebles = inmuebles.rename(columns=columnas)
inmuebles.sample(10)
```

```python
type (inmuebles.Valor[0]) #Arroja valor str (cadena de caracteres) se identifica porque aparece entre comillas
```

```python
inmuebles.Valor[0]
```

```python
inmuebles.Valor[0]+inmuebles.Valor[1] #para sumar 2 objetos 
#para escribir un rango uso :(numero hasta el que quiero llegar)
```

```python
inmuebles.Valor[0].split() #split: funcion para separar una cadena de texto
```

```python
valor = inmuebles.Valor.str.split(expand=True) #inmuebles.Valor=panda series, str=devolver lista Expand=True ¨: Devuelve un data frame
inmuebles["Moneda"] = valor[0]
inmuebles["Precio"] = valor[1]
inmuebles.sample(3)
```

```python
inmuebles.info()
```

```python
inmuebles["Precio"] = inmuebles["Precio"].str.replace('.','',regex=True) #replace: sive para reemplazar siepre debe ir acompañada de regex=True
```

```python
inmuebles[["Precio","Barrio"]] 
```

```python
inmuebles["Precio_Millon"] = inmuebles.Precio.astype("float")/1000000
inmuebles.info()
```

```python
inmuebles.describe()
```

```python
pd.set_option("display.precision",2) 
pd.set_option("display.float_format",lambda x:"%.2f" % x) #Eliminar posiciones decimales
inmuebles.describe()
```
